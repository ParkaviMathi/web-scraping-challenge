```python
#Importing Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pymongo
from pprint import pprint
```

```python
# Settingup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Initialise PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.mars_db
collection = db.articles
```

```python
#Setting up browser for the Website from which the data is to be scraped
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
# Iterate through all pages
for x in range(15):
    # HTML object
    html = browser.html
    # Parse HTML with Beautiful Soup
    soup = BeautifulSoup(html, 'html.parser')
    # Retrieve all elements that contain book information
    articles = soup.find_all('div', class_='list_text')
    
    # Iterate through each article
    for article in articles:
        # scrape the article header 
        title = article.find('div', class_='content_title').text.strip()
        
        # scrape the article content preview
        preview = article.find('div', class_='article_teaser_body').text.strip()
        
        print('-----------')
        print(title)
        print(preview)
        
        
        # Dictionary to be inserted into MongoDB
        post = {
            'title': title,
            'preview': preview,
        }

        # Insert dictionary into MongoDB as a document
        collection.insert_one(post)
```

```python
# Display the MongoDB records created above as a Python list
ls_dict=[]
items = db.articles.find()
for item in items:
    ls_dict.append(item)
    
pprint(ls_dict)
```

```python
browser.quit()

```
